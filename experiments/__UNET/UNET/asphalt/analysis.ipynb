```python
from torch.utils.data import Dataset, DataLoader
import os, cv2
import numpy as np
from PIL import Image
import torch
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

```python
to_pickle = []

for file in test_files:
    file = file.strip()
    path = os.path.join('/projects/0/gusr51794/srst_scratch_drive/binary_training/baseline_test_images/512/', file.strip())

    print(path)

    img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)

    img_transforms = transforms.Compose([
        transforms.ToPILImage(),
        transforms.ToTensor(), # convert the image to a tensor
    ])
    img = img_transforms(img)
    img = img.unsqueeze(0) # add batch dimension to the image tensor
    img_np = img.squeeze(0).cpu().numpy().transpose((1, 2, 0))

    
    to_pickle.append({
        'img_np': img_np,
        'img': img,
        'path': path
    })
```

```python
import pickle
```

```python
# Save pickle

with open(f'{ANALYSIS_DIR}/test_images.pickle', 'wb') as f:
    pickle.dump(to_pickle, f)
```

```python
# Load pickle
with open(f'{ANALYSIS_DIR}/test_images.pickle', 'rb') as f:
    test_images_pickle = pickle.load(f)
```

```python
len(test_images_pickle)
```

```python
for img in test_images_pickle:
    # Remove the batch dimension, convert to numpy, and transpose to (height, width, channels)
    img_np = img['img']
    img_np = img_np.squeeze(0).cpu().numpy().transpose((1, 2, 0))

    plt.imshow(img_np)
    plt.show()
```

```python
img_predictions = []
for img_to_pred in test_images_pickle:
    img = img_to_pred['img']
    img_np = img_to_pred['img_np']
    with torch.no_grad():
        output = model(img)
        
        img_predictions.append({
            'img_name': 'test_image',
            'img': img_to_pred,
            'output': output
        })
        
        
        # Move to CPU and convert to numpy
        output_np = output.cpu().numpy()
        output_binary_np = output.squeeze(0).squeeze(0).cpu().numpy()  # Squeeze the channel dimension
        
        
                # Convert to binary by thresholding
        output_binary = output_binary_np < 0.5


        # Plot the binary image
        plt.subplot(1, 2, 2)
        plt.imshow(output_binary, cmap='binary')
        plt.title('Predicted Image')

        plt.imshow(img_np, cmap='binary')


```

```python
for inf in img_predictions:
    img_np = inf['img']['img_np']
    output_binary = inf['output'].squeeze(0).squeeze(0).cpu().numpy()
    plt.subplot(1, 2, 1)
    plt.imshow(img_np)
    plt.title('Original Image')
    plt.subplot(1, 2, 2)
    plt.imshow(output_binary <  -2, cmap='binary')
    plt.title('Predicted Image')
    plt.show()
```
